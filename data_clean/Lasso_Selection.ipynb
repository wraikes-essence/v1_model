```python
def AIC(model, X, y):
    k = X.shape[1]
    y_hat = model.predict(X)
    
    resid = y - y_hat
    sse = sum(resid**2)
    aic = 2*k - 2*np.log(sse)
    
    return aic
```

```python
sc = StandardScaler()

X = df.drop(columns='answer_desired_Aided_awareness')
y = df['answer_desired_Aided_awareness']

C = [0.01, 0.05, 0.1]
aic_scores = []

for c in C:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    
    lasso = LogisticRegression(penalty='l1', random_state=4, C=c)
    lasso.fit(X_train, y_train)
    aic = AIC(lasso, X_test, y_test)
    aic_scores.append(aic)   
```
